### 3-1.Setup HSV space threshold (15%)

```python
'''
#You should find the config file 'universal.yaml'
hsv_white1 = ???
hsv_white2 = ???
hsv_yellow1 = ???
hsv_yellow2 = ???
hsv_red1 = ???
hsv_red2 = ???
hsv_red3 = ???
hsv_red4 = ???
'''

hsv_white1 = np.array([0,0,150])
hsv_white2 =  np.array([180,50,255])
hsv_yellow1 = np.array([20,120,50])
hsv_yellow2 = np.array([40,255,255])
hsv_red1 =    np.array([0,100,100])
hsv_red2 =    np.array([15,255,255])
hsv_red3 =    np.array([165,100,100])
hsv_red4 =    np.array([180,255,255])
```

### 3-2.Threshold colors in HSV space (15%)

```python
'''
#change color space to HSV
hsv = ???

find the color
white = ???
kernel = ???
white = ???

yellow = ???
kernel = ???
yellow = ???

red1 = ???
red2 = ???
red = ???
kernel = ???
red = ???
'''
#change color space to HSV
hsv = cv2.cvtColor(image_cv, cv2.COLOR_BGR2HSV)

#find the color
white = cv2.inRange(hsv, hsv_white1, hsv_white2)
#kernel = np.ones((3,3),np.float32)/9
#white = cv2.filter2D(white,-1,kernel)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
white = cv2.dilate(white, kernel)

yellow = cv2.inRange(hsv, hsv_yellow1, hsv_yellow2)
#kernel = np.ones((3,3),np.float32)/9
#yellow = cv2.filter2D(yellow,-1,kernel)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
yellow = cv2.dilate(yellow, kernel)

red1 = cv2.inRange(hsv, hsv_red1, hsv_red2)
red2 = cv2.inRange(hsv, hsv_red3, hsv_red4)
red = cv2.bitwise_or(red1, red2)
#kernel = np.ones((3,3),np.float32)/9
#red = cv2.filter2D(red,-1,kernel)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
red = cv2.dilate(red, kernel)

# Uncomment '#' to plot with color
x = cv2.cvtColor(yellow, cv2.COLOR_GRAY2BGR)
x[:,:,2] *= 1
x[:,:,1] *= 1
x[:,:,0] *= 0
x = cv2.cvtColor(x, cv2.COLOR_BGR2RGB)

y = cv2.cvtColor(red, cv2.COLOR_GRAY2BGR)
y[:,:,2] *= 1
y[:,:,1] *= 0
y[:,:,0] *= 0
y = cv2.cvtColor(y, cv2.COLOR_BGR2RGB)

plt.subplot(131),plt.imshow(white,cmap = 'gray')
plt.title('White'), plt.xticks([]), plt.yticks([])

#plt.subplot(132),plt.imshow(yellow,cmap = 'gray')
plt.subplot(132),plt.imshow(x,cmap = 'brg')
plt.title('Yellow'), plt.xticks([]), plt.yticks([])

#plt.subplot(133),plt.imshow(red,cmap = 'gray')
plt.subplot(133),plt.imshow(y,cmap = 'brg')
plt.title('Red'), plt.xticks([]), plt.yticks([])

plt.show()
```

### 4.Combine Edge and Colors (15%)

```python
'''
edge_color_white = ???
edge_color_yellow = ???
edge_color_red = ???
'''
#get edges
edge_color_w = cv2.Canny(white,60,150) 
edge_color_y = cv2.Canny(x,60,150)
edge_color_r = cv2.Canny(y,60,150)

#refine edges
edge_color_white = cv2.bitwise_and(white, edge_color_w)
edge_color_yellow = cv2.bitwise_and(yellow, edge_color_y)
edge_color_red = cv2.bitwise_and(red, edge_color_r)

#change color
edge_color_white = cv2.cvtColor(edge_color_white, cv2.COLOR_GRAY2BGR)

edge_color_yellow = cv2.cvtColor(edge_color_yellow, cv2.COLOR_GRAY2BGR)
edge_color_yellow[:,:,2] *= 1
edge_color_yellow[:,:,1] *= 1
edge_color_yellow[:,:,0] *= 0
edge_color_yellow = cv2.cvtColor(edge_color_yellow, cv2.COLOR_BGR2RGB)

edge_color_red = cv2.cvtColor(edge_color_red, cv2.COLOR_GRAY2BGR)
edge_color_red[:,:,2] *= 1
edge_color_red[:,:,1] *= 0
edge_color_red[:,:,0] *= 0
edge_color_red = cv2.cvtColor(edge_color_red, cv2.COLOR_BGR2RGB)

#combine edges and color
edge_color = edge_color_red + edge_color_yellow + edge_color_white


plt.subplot(141),plt.imshow(edge_color_white,cmap = 'gray')
plt.title('Edge Color W'), plt.xticks([]), plt.yticks([])
plt.subplot(142),plt.imshow(edge_color_yellow,cmap = 'brg')
plt.title('Edge Color Y'), plt.xticks([]), plt.yticks([])
plt.subplot(143),plt.imshow(edge_color_red,cmap = 'brg')
plt.title('Edge Color R'), plt.xticks([]), plt.yticks([])
plt.subplot(144),plt.imshow(edge_color,cmap = 'brg')
plt.title('Edge Colors'), plt.xticks([]), plt.yticks([])

plt.show()
```
